This describes the assigning of a reference indexing value to certain age groups.

```python
dataset = [file1]
for data in dataset:
    data.loc[data['age'] <= 18, 'age'] = 0
    data.loc[(data['age'] > 18) & (data['age'] <= 32),'age'] = 1
    data.loc[(data['age'] > 32) & (data['age'] <= 46),'age'] = 2
    data.loc[(data['age'] > 46) & (data['age'] <= 64),'age'] = 3
    data.loc[data['age'] > 64, 'age'] = 4
```
